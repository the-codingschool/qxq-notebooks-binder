#### **Problem #6.4.8**

Change the value of the `age` variable such that the final price is $90.

<br>

**NOTE**: Although the code involves several new and complex concepts, try reading the comments to understand what the code does.

````python
age = 5  # This is the age of the person
price = 100  # The original price of the item

# If the person is 18 or older, they qualify for a 10% discount
if age >= 18:
    discount = price * 0.10  # Calculate 10% of the price
    final_price = price - discount  # Subtract the discount from the original price
    print("You qualify for a discount!")

# If the person is younger than 18, they don't qualify for a discount
else:
    final_price = price  # The original price remains the same
    print("No discount available.")


print("The final price is:", final_price)
````

##### **Solution**

````python
age = 18  # This is the age of the person
price = 100  # The original price of the item

# If the person is 18 or older, they qualify for a 10% discount
if age >= 18:
    discount = price * 0.10  # Calculate 10% of the price
    final_price = price - discount  # Subtract the discount from the original price
    print("You qualify for a discount!")

# If the person is younger than 18, they don't qualify for a discount
else:
    final_price = price  # The original price remains the same
    print("No discount available.")


print("The final price is:", final_price)
````
